2. Создаем объект DataFrame на основе 3-х объектов Series. Количество строк – 24

```python
data = {'A': pd.Series(np.random.randn(24)),
        'B': pd.Series(np.random.randn(24)),
        'C': pd.Series(np.random.randn(24))}
df = pd.DataFrame(data)
```

 3. Выводим информацию о количестве строк в таблице

4. Выбираем строки с 12 по 17 и столбцы 2 и 3. Реализовываем двумя способами: а) индикатор loc, б) индикатор iloc

```python
print("Выборка с помощью loc:")
df.loc[11:16, ['B', 'C']]
print("Выборка с помощью iloc:")
df.iloc[11:17, 1:3]
```
