```python
# Создание функции для агрегирования данных
def get_purchase_info(products):
    num_purchases = len(products)
    unique_products = products.unique()
    return pd.Series([unique_products, num_purchases], index=['product_names', 'num_purchases'])
# Группировка данных по клиентам и применение функции agg()
purch = df.groupby('customer_id')['product_name'].agg(get_purchase_info).reset_index()
# Разделение столбца 'product_name' на два столбца 'product_names' и 'num_purchases'
purch[['product_names', 'num_purchases']] = pd.DataFrame(purch['product_name'].tolist(), index=purch.index)
# Удаление столбца 'product_name'
purch.drop('product_name', axis=1, inplace=True)
# Вывод таблицы 'purch'
purch
```

Добавить в таблицу df столбец 'store_sales_group'
с категорией суммы покупки:
- `small` - сумма покупки < 10;
- `average` - сумма покупки >= 10 и < 13;
- `large` - сумма покупки >= 13.

Использовать функцию `applay()`.

```python
# Функция для определения категории суммы покупки
def categorize_sales_amount(sales_amount):
    if sales_amount < 10:
        return 'small'
    elif sales_amount >= 10 and sales_amount < 13:
        return 'average'
    else:
        return 'large'
# Добавление нового столбца 'store_sales_group' с помощью apply()
df['store_sales_group'] = df['store_sales'].apply(categorize_sales_amount)
# Вывод обновленной таблицы df
df
```

Для выполнения следующих заданий используется набор данных о динамике велосипедного движения между центром города и некоторым его районом:
- `Time` - время измерения;
- `direction_centre`	- количество велосипедистов в направлении центра;
- `direction_region` - количество велосипедистов в направлении района.

Считать файл 'time_series.csv' и записать его в таблицу 'dft'.

```python
dft = pd.read_csv('time_series.csv') #считывание файла в переменную dft
dft.head() #проверка правильности считывания
```

На основе информаци о дате и времени создать в таблице dft столбец 'Date_Time'.

Элемент нового столбца должен иметь следующий вид:
14/07/2011	17:44

```python
#Создаем столбец 'Date_Time' в формате 'dd/mm/yyyy hh:mm'
dft['Date_Time'] = dft['Date'] + ' ' + dft['Time']
```

Удалить столбцы 'Date' и 'Time'.

```python
# Удаляем столбцы 'Date' и 'Time'
dft.drop(['Date', 'Time'], axis=1, inplace=True)
```

Преобразовать тип переменной 'Date_Time' в формат `datetime`.


```python
# Преобразуем тип переменной 'Date_Time' в формат datetime
dft['Date_Time'] = pd.to_datetime(dft['Date_Time'], format='%d/%m/%Y %H:%M')
```

Установить столбец 'Date_Time' в качестве индекса.

```python
# Устанавливаем столбец 'Date_Time' в качестве индекса
dft.set_index('Date_Time', inplace=True)
```

Добавить признак 'both' с суммарным потоком велосипедистов.

```python
# Добавляем признак 'both' с суммарным потоком велосипедистов
dft['both'] = dft['direction_centre'] + dft['direction_region']
```

Добавить столбец 'weekday' с названиями дней недели на русском языке.

```python
import calendar
# Создаем функцию для преобразования в русский язык числовых значений
def get_russian_weekday(day):
  russian_weekdays = ['Понедельник', 'Вторник', 'Среда', 'Четверг', 'Пятница', 'Суббота', 'Воскресенье']
  return russian_weekdays[day]

# Добавляем столбец с русскими названиями
dft['weekday'] = dft.index.dayofweek.map(get_russian_weekday)
dft
```

Выбрать временные интервалы, для которых суммарный поток менее 3 велосипедистов.

```python
#  Выбираем временные интервалы, для которых суммарный поток менее 3 велосипедистов
dft[dft['both'] < 3]
```

Выбрать временные интервалы, для которых в одном из направлений проехало менее 5 велосипедистов.

```python
# Выбираем временные интервалы, для которых в одном из направлений проехало менее 5 велосипедистов
dft[(dft['direction_centre'] < 5) | (dft['direction_region'] < 5)]
```

Определить среднее количество велосипедистов за 1 час.

```python
# Определяем среднее количество велосипедистов за 1 час
dft.resample('1H').sum().mean()
```

Найти даты с 5-ю самыми большими значениями количества велосипедистов, проехавших в направлении к центру.

```python
# Находим даты с 5-ю самыми большими значениями количества велосипедистов, проехавших в направлении к центру
dft.nlargest(5, 'direction_centre').index
```
