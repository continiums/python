# Задача 1.
С использованием метода `plot()` написать код для построения графика линии.

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 10)
y = x * 2

plt.plot(x, y, color="green", linestyle="--", marker="o", markersize=7, markerfacecolor="red")
plt.title("График линии", color = "purple")
plt.xlabel("Ось X", color="blue")
plt.ylabel("Ось Y", color="blue")
plt.grid(True)

# Установка частоты делений по оси x
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))

# Установка нулей графика
plt.xticks(np.arange(0, 10, 1))
plt.yticks(np.arange(0, 19, 2))

plt.show()
```

# Задача 2.
С использованием метода `plot()` написать код для построения двух графиков линии в одной координатной сетке.

image.png

```python
x = np.arange(1, 10)
y1 = x * 2
y2 = x * 2 + 3

plt.plot(x, y1, color="green", linestyle="--", marker="o", markersize=7, markerfacecolor="red", label = "функция f(x) = 2 * x")
plt.plot(x, y2, color="blue", linestyle=":", marker="s", markersize=7, markerfacecolor="green", label = "функция f(x) = 2 * x + 3")
plt.title("Графики линейных функций", color = "purple")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.grid(True)
plt.yticks(np.arange(0, 24, 1))
plt.legend()
plt.show()
```

# Задача 3.
С использованием метода `subplots()` написать код для построения 4-х графиков линии в одной одном окне в разных координатных сетках.

image.png

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 10)
y1 = x * 2
y2 = x * 2 + 1
y3 = x * 2 + 2
y4 = x * 2 + 3

fig, axs = plt.subplots(2, 2)
fig.suptitle('Графики линий')

axs[0, 0].plot(x, y1, color='red')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].set_title('f(x)*2+0')
axs[0, 0].grid(True, which='major', linestyle='--')  # Увеличение размера сетки

axs[0, 1].plot(x, y2, color='blue')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')
axs[0, 1].set_title('f(x)*2+1')
axs[0, 1].grid(True, which='both', linestyle='--')  # Увеличение размера сетки

axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')
axs[1, 0].set_title('f(x)*2+2')
axs[1, 0].grid(True, which='both', linestyle='--')  # Увеличение размера сетки

axs[1, 1].plot(x, y4, color='purple')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')
axs[1, 1].set_title('f(x)*2+3')
axs[1, 1].grid(True, which='both', linestyle='--')  # Увеличение размера сетки

plt.tight_layout()
plt.show()
```

# Задача 5.
Построить столбчатую диаграмму, отображающую относительное количество атомобилей с различными трансмиссиями (`transmission`) с использованием:
- функции `plt.bar()`;
- функций `DataFrame.plot()` или `DataFrame.plot.bar()`.

image.png

```python
import matplotlib.pyplot as plt
import pandas as pd

# Создаем пример данных
data = {'тип трансмиссии': ['Manual', 'Automatic'],
        'доля автомобилей': [0.9, 0.17]}  # Пример значений количества

# Преобразуем данные в DataFrame
df = pd.DataFrame(data)

# Построение столбчатой диаграммы с помощью plt.bar()
plt.figure(figsize=(8, 6))
bars = plt.bar(df['тип трансмиссии'], df['доля автомобилей'], color=['red', 'blue'], edgecolor='black', linewidth=2)

plt.xlabel('тип трансмиссии')
plt.ylabel('доля автомобилей')
plt.title('распределение автомобилей')
plt.show()
```

```python
# Построение столбчатой диаграммы с помощью DataFrame.plot.bar()
plt.figure(figsize=(8, 6))
df.plot.bar(x='тип трансмиссии', y='доля автомобилей', color=['red', 'blue'],  edgecolor='black', linewidth=2, legend=None, width=1)
plt.xlabel('Transmission')
plt.ylabel('доля автомобилей')
plt.title('распределение автомобилей')
plt.show()
```
